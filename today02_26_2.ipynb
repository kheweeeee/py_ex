```python
sample = "abc123def56"

re.findall(r"\d+", sample)

re.findall(r"[a-z]+", sample)
```

```python
# 특정 형식 찾기

data = "test12@gmail.com"

email_pattern = r"\w+@\w+\.\w+"
# \w 문자 또는 숫자
# +: 1개 이상
# @: 말 그대로 @
# \.: 점(특수 문자이기 때문에 \를 붙인다)

emails = re.findall(email_pattern, data)
print(emails)
```

```python
# 문자열 치환

test = "가격은 1200원 입니다."

new_text = re.sub(r"\d+", "", test)
print(new_text)
```

```python
# 정규식과 일반 문자열 처리

test = "몸무게:56kg, 키:180cm"

test.split(",")
number = re.findall(r"\d+", test)
print(number)
```
