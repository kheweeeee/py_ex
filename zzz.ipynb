```python
# 빈 클래스 만들기
class person:
    def hello(self):
        print("hello")

    def hi(self):
        self.hello()  # 클래스 내에서 다른 메서드 호출

jin = person()
jin.hi()
```

```python
print("== 정답 v1 ==")
month_end_days = {}

month_end_days["1월"] = 31
month_end_days["2월"] = 28
month_end_days["3월"] = 31

print(month_end_days)

print("== 정답 v2 ==")
month_end_days = {}

for i in range(1, 12 + 1):
  month = str(i) + "월"

  end_day = 31

  if i == 2:
    end_day = 28
  elif i in [4, 6, 9, 11]:
    end_day = 30

  month_end_days[month] = end_day

print(month_end_days)

print("== 정답 v3 ==")
month_end_days = {}

month_end_days_list_ver = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

for i, end_day in enumerate(month_end_days_list_ver):
  month = str(i + 1) + "월"
  month_end_days[month] = end_day

print(month_end_days)
```
